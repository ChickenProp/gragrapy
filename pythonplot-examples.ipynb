```python
%matplotlib inline
import gragrapy as gg
```

```python
gg.plot(gg.data.mpg).aes(x='manufacturer').geom('bar', size=20).coord('flip').title('Number of cars by make')
```

```python
gg.plot(gg.data.mpg).aes(x='cty').geom('hist', binwidth=2)
```

```python
(gg.plot(gg.data.mpg)
   .aes(x='displ', y='hwy')
   .geom('point')
   .labels(x='Engine displacement in liters',
           y='Highway MPG',
           title='Engine displacement in liters vs highway MPG'))
```

```python
import pandas as pd
import numpy as np

ts = (pd
      .Series(np.random.randn(1000), 
              index=pd.date_range('1/1/2000', periods=1000))
      .cumsum()
      .to_frame()
      .reset_index()
     )
ts.columns = ['date', 'value']

gg.plot(ts).aes('date', 'value').geom('line')
```

```python
(gg.plot(gg.data.mpg)
   .aes(x='displ', y='hwy', color='class')
   .geom('point')
   .labels(x='Engine displacement in liters',
           y='Highway MPG',
           title='Engine displacement in liters vs highway MPG'))
```

```python
gg.plot(gg.data.mpg).aes(x='cty', y='hwy', size='cyl').geom('point', alpha=.5)
```

```python
# looks like you can't facet on python keywords
gg.plot(gg.data.mpg.assign(c=gg.data.mpg['class'])).aes(x='displ', y='hwy').geom('point').facet('c')
```

```python
gg.plot(gg.data.mpg).aes(x='displ', y='hwy').geom('point').facet_grid(('drv', 'cyl'))
```

```python
gg.plot(gg.data.mpg).aes('displ', 'hwy').geom('point').geom('smooth', method='lm')
```

```python
(gg.plot(gg.data.mpg)
   .aes(x='displ', y='hwy')
   .geom('point', gg.aes(color='class'))
   .geom('smooth', data=gg.data.mpg[gg.data.mpg['class'] == 'subcompact'], se=False, method='loess'))
```
